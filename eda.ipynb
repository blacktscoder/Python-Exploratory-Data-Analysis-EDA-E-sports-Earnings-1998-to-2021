```python
corr = dataset.corr()
sns.heatmap(corr, xticklabels = corr.columns, yticklabels = corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

What we can conclude here is that as older the ReleaseDate as lower is TotalEarnings, OnlineEarnings, TotalPlayers and TotalTournamens. This means that in the past they didnt have many players or championships, so the gains were very low or almost nil.

```python
fig, ax = plt.subplots(figsize = (18,10))
ax.scatter(dataset['TotalTournaments'], dataset['ReleaseDate'])

# X-axis label
ax.set_xlabel('Tournaments')

# y-axis label
ax.set_ylabel('Per Year')
plt.show()
```
